```python
data_lst_result = json_data['results']
data_lst_result
```

```python
print(type(data_lst_result))
print(len(data_lst_result))
```

```python
score_rank_lst = list()
for item in data_lst_result:
    score_rank_lst.append([item['rank'], item['score']])
score_rank_lst[:20]
```

```python
df = pd.DataFrame(score_rank_lst, columns=['rank','score'])
df.head()
```
